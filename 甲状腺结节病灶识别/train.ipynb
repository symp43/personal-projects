```python
import os
import cv2
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.transforms as transforms
import json
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from pathlib import Path
from sklearn.model_selection import train_test_split
from torch.optim.lr_scheduler import CosineAnnealingLR, LinearLR, LambdaLR, SequentialLR
from torchvision.models import resnet34
from torch.utils.data import Dataset, DataLoader

# 模块化导入
from datasets import MedicalDataset, DualTransform
from model import AttentionUNetResNet34
from metrics import dice_coeff, iou_score, visualize_data, plot_prediction, DiceFocalLoss, StabilizedBCE, DiceLoss, PrecisionBalancedLoss

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 配置参数
DATA_ROOT = Path(r"C:\Personal\Desktop\tn3k")  # 数据集根目录
BATCH_SIZE = 4
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
use_amp = True  # 启用混合精度训练
PATIENCE = 10  # 容忍多少个 epoch 没有提升
JSON_FOLDS = [  # JSON文件路径
    "tn3k-trainval-fold0.json"
]

# 定义三阶段训练配置
class TrainingConfig:
    def __init__(self):
        self.stages = {
            1: {"epochs": 5, "lr_encoder": 1e-6, "lr_decoder": 1e-5, "loss": "StabilizedBCE"},
            2: {"epochs": 5, "lr_encoder": 1e-6, "lr_decoder": 1e-5, "loss": "DiceLoss"},
            3: {"epochs": 10, "lr_encoder": 5e-6, "lr_decoder": 5e-5, "loss": "DiceFocalLoss"}
        }
        self.current_stage = 1
        self.start_epoch = 0
        self.total_epochs = sum(stage["epochs"] for stage in self.stages.values())
    
    def get_stage(self, epoch):
        #根据当前epoch确定训练阶段
        cumulative = 0
        for stage, config in self.stages.items():
            cumulative += config["epochs"]
            if epoch < cumulative:
                return stage, epoch - (cumulative - config["epochs"])
        return 3, epoch  

# 数据加载与准备
train_transform = DualTransform(p=0.8)  # 训练集增强

# 完整训练集（用于交叉验证）
full_train_dataset = MedicalDataset(
    DATA_ROOT / "trainval-image",
    DATA_ROOT / "trainval-mask",
    transform=train_transform,
    stage=1,
    mode="train"
)

test_dataset = MedicalDataset(
    DATA_ROOT / "test-image",
    DATA_ROOT / "test-mask",
    mode="test"
)

test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=0)

def custom_collate_fn(batch):
    images = [item[0] for item in batch]
    masks = [item[1] for item in batch]
    
    # 合并图像
    images_tensor = torch.stack(images, dim=0)
    
    # 合并掩码并转换为 float32
    masks_tensor = torch.stack(masks, dim=0).to(torch.float32)
    
    return images_tensor, masks_tensor
    
# 交叉验证结果存储
cv_results = []


def overlay_prediction(image_tensor, mask_tensor, pred_tensor, title="预测vs真实"):
    image = image_tensor.squeeze().cpu().numpy()
    mask = mask_tensor.squeeze().cpu().numpy()
    pred = pred_tensor.squeeze().cpu().numpy()

    fig, ax = plt.subplots(figsize=(6,6))
    ax.imshow(image, cmap='gray')
    ax.contour(mask, colors='g', linewidths=1.5, label='GT')       # 绿色是真实结节
    ax.contour(pred, colors='r', linewidths=1.5, label='Pred')     # 红色是预测边界
    ax.set_title(title)
    plt.axis('off')
    plt.show()

def find_best_threshold(preds, targets, thresholds=None):
    if thresholds is None:
        thresholds = torch.arange(0.3, 0.81, 0.05)

    best_thresh = 0.5
    best_dice = 0.0

    for t in thresholds:
        pred_bin = (preds > t).float()
        dice = dice_coeff(pred_bin, targets)
        if dice > best_dice:
            best_dice = dice
            best_thresh = t.item() if isinstance(t, torch.Tensor) else t

    return best_thresh, best_dice 

def get_loss_function(name, device):
    if name == "StabilizedBCE":
        return StabilizedBCE(pos_weight=torch.tensor([10.0], device=device))
    elif name == "DiceLoss":
        return DiceLoss()
    elif name == "DiceFocalLoss":
        return DiceFocalLoss()
    else:
        raise ValueError(f"未知损失函数: {name}")    


# 创建优化器函数
def create_optimizer(model, stage_config):
    encoder_params = []
    decoder_params = []
    
    # 分离编码器和解码器参数
    for name, param in model.named_parameters():
        if "encoder" in name or "resnet" in name:
            encoder_params.append(param)
        else:
            decoder_params.append(param)
    
    optimizer = torch.optim.AdamW([
        {'params': encoder_params, 'lr': stage_config["lr_encoder"]},
        {'params': decoder_params, 'lr': stage_config["lr_decoder"]}
    ], weight_decay=1e-6)
    
    return optimizer

# 创建数据集函数
def create_dataset_subset(full_dataset, indices, stage):
    subset = torch.utils.data.Subset(full_dataset, indices)
    
    subset.dataset.stage = stage
   
    return subset

def evaluate_model(model, test_loader, best_thresh, device):
    """在测试集上评估模型性能"""
    model.eval()
    test_dice = 0.0
    test_iou = 0.0
    test_preds = []
    test_targets = []
    
    # 存储可视化样本
    visual_samples = []
    
    with torch.no_grad():
        for i, (images, masks) in enumerate(test_loader):
            images = images.to(device)
            masks = masks.to(device)
            
            outputs = model(images)
            outputs = torch.sigmoid(outputs)
            
            # 使用最佳阈值进行二值化
            preds = (outputs > best_thresh).float()
            
            # 计算指标
            test_dice += dice_coeff(preds, masks).item()
            test_iou += iou_score(preds, masks).item()
            
            # 收集前3个样本用于可视化
            if i == 0 and len(visual_samples) < 3:
                for j in range(min(3, images.size(0))):
                    visual_samples.append((
                        images[j].cpu().squeeze().numpy(),
                        masks[j].cpu().squeeze().numpy(),
                        preds[j].cpu().squeeze().numpy()
                    ))
    
    # 计算平均指标
    test_dice /= len(test_loader)
    test_iou /= len(test_loader)
    
    return test_dice, test_iou, visual_samples

# 交叉验证主循环
for fold, json_path in enumerate(JSON_FOLDS):
    print(f"\n===== 开始第 {fold+1}/{len(JSON_FOLDS)} 折交叉验证 =====")
    
    # 读取JSON文件获取训练/验证索引
    with open(json_path, 'r') as f:
        fold_data = json.load(f)
    train_indices = fold_data['train']
    val_indices = fold_data['val']
    
    # 初始化训练配置
    config = TrainingConfig()
    
    # 创建模型
    model = AttentionUNetResNet34(in_channels=1, out_channels=1, pretrained=True).to(DEVICE)
    
    # 初始冻结编码器
    model.freeze_all()
    
    # 获取第一阶段的配置
    stage_config = config.stages[config.current_stage]
    
    # 创建优化器
    optimizer = create_optimizer(model, stage_config)
    
    # 初始化损失函数
    criterion = StabilizedBCE(pos_weight=10.0)
    scaler = torch.amp.GradScaler(enabled=use_amp)
    
    # 创建数据集子集
    train_subset = create_dataset_subset(full_train_dataset, train_indices, config.current_stage)
    val_subset = create_dataset_subset(full_train_dataset, val_indices, config.current_stage)
    
    # 创建数据加载器 
    train_loader = DataLoader(
        train_subset,
        batch_size=BATCH_SIZE,
        shuffle=True,
        num_workers=0,
        collate_fn=custom_collate_fn 
    )
    
    val_loader = DataLoader(
        val_subset,
        batch_size=BATCH_SIZE,
        shuffle=False,
        num_workers=0,
        collate_fn=custom_collate_fn  
    )
    
   
    # 早停机制初始化
    best_dice = 0.0
    patience_counter = 0
    fold_checkpoint_path = f"best_model_fold{fold}.pth"
    
    warmup_scheduler = None
    warmup_start_epoch = 0
    warmup_epochs = 3  # 初始预热epoch数
    
    loss_transition_alpha = 0.0
    old_criterion = None
    new_criterion = None
    
    # 训练循环
    print(f"开始第 {fold+1} 折训练 - 设备: {DEVICE}, 批次大小: {BATCH_SIZE}, 混合精度: {use_amp}")
    
    for epoch in range(config.total_epochs):
        # 确定当前阶段和阶段内epoch
        stage, stage_epoch = config.get_stage(epoch)
        
        # 阶段切换处理
        if stage != config.current_stage:
            print(f"\n=== 切换到阶段 {stage}/3 ===")
            config.current_stage = stage
            stage_config = config.stages[stage]
            
            # 更新模型冻结状态
            if stage == 1:
                model.unfreeze_from()  # 只解冻解码器
            elif stage == 2:
                model.unfreeze_from("layer4")
                print("解冻layer4")
            elif stage == 3:
                model.unfreeze_all()
                print("完全解冻模型")
            # 保存旧损失函数
            old_criterion = criterion
    
            # 更新损失函数
            if stage_config["loss"] == "PrecisionBalancedLoss":
                criterion = PrecisionBalancedLoss()
            elif stage_config["loss"] == "FinalHybridLoss":
                criterion = FinalHybridLoss()
            print(f"使用损失函数: {stage_config['loss']}")
            
            loss_transition_alpha = 0.0
            
            # 更新优化器学习率
            for i, param_group in enumerate(optimizer.param_groups):
                if i == 0:  # 编码器参数
                    param_group['lr'] = stage_config["lr_encoder"]
                else:  # 解码器参数
                    param_group['lr'] = stage_config["lr_decoder"]
            print(f"更新学习率 - 编码器: {stage_config['lr_encoder']:.1e}, 解码器: {stage_config['lr_decoder']:.1e}")
            
            scaler = torch.amp.GradScaler(enabled=use_amp)
            
            # 阶段切换后的学习率预热
            print("应用阶段切换后的学习率预热")
            
            # 保存目标学习率
            target_lrs = [group['lr'] for group in optimizer.param_groups]
        
            # 暂时降低学习率
            for i, group in enumerate(optimizer.param_groups):
                group['lr'] = target_lrs[i] * 0.1  # 降低到目标值的10%
                
            # 创建新的预热调度器
            warmup_scheduler = torch.optim.lr_scheduler.LambdaLR(
                optimizer, 
                lr_lambda=lambda epoch_idx: 0.1 + 0.9 * min(1.0, (epoch_idx + 1) / warmup_epochs)  
            )
            warmup_start_epoch = epoch  
            
            # 更新数据集的阶段
            train_subset.dataset.stage = stage
            val_subset.dataset.stage = stage
            
        # 应用学习率预热
        if warmup_scheduler is not None:
            # 更新学习率
            warmup_scheduler.step()

            # 计算预热进度
            warmup_progress = min(1.0, (epoch - warmup_start_epoch + 1) / warmup_epochs)
            print(f"学习率预热进度: {warmup_progress*100:.0f}%")
        
            # 检查预热是否结束
            if epoch - warmup_start_epoch >= warmup_epochs - 1:
                warmup_scheduler = None
                print(f"阶段 {stage} 学习率预热结束")
         
                # 恢复学习率为 stage 对应的目标值
                for i, param_group in enumerate(optimizer.param_groups):
                    param_group['lr'] = target_lrs[i]   
                    
        # 打印阶段信息
        print(f"\n=== 阶段 {stage}/3 | 阶段内epoch {stage_epoch+1}/{stage_config['epochs']} ===")
        print(f"训练集大小: {len(train_subset)}, 验证集大小: {len(val_subset)}")
        
        # 训练阶段
        model.train()
        train_loss = 0.0
        progress_bar = tqdm(enumerate(train_loader), total=len(train_loader))
        grad_means = []
        grad_maxs = []
        
        for i, (images, masks) in progress_bar:
            images = images.to(DEVICE, non_blocking=True)
            masks = masks.to(DEVICE, non_blocking=True)
            
            optimizer.zero_grad()
            
            with torch.amp.autocast(device_type='cuda', enabled=use_amp):
                images = images.to(torch.float16 if use_amp else torch.float32)
                outputs = model(images)
                loss = criterion(outputs, masks)
            
            # 损失函数过渡
            if old_criterion is not None and new_criterion is not None:
                loss_old = old_criterion(outputs, masks)
                loss_new = new_criterion(outputs, masks)
                loss = loss_transition_alpha * loss_new + (1 - loss_transition_alpha) * loss_old
                
            else:
                loss = criterion(outputs, masks)
                
            scaler.scale(loss).backward()   
            # 梯度监控
            grad_norms = []
            for name, param in model.named_parameters():
                if param.requires_grad and param.grad is not None:
                    grad_norms.append(param.grad.abs().mean().item())
                    
            grad_mean = np.mean(grad_norms)
            grad_max = np.max(grad_norms) if grad_norms else 0
            grad_means.append(grad_mean)
            grad_maxs.append(grad_max)
        
            # 梯度裁剪
            scaler.unscale_(optimizer)
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=0.5)
            
            scaler.step(optimizer)
            scaler.update()
            
            train_loss += loss.item() * images.size(0)
            progress_bar.set_description(f"Epoch {epoch+1}/{config.total_epochs}, 批次 {i+1}/{len(train_loader)}, 损失: {loss.item():.4f}")
            progress_bar.set_postfix({
                'grad_mean': f"{np.mean(grad_norms):.1e}",
                'grad_max': f"{np.max(grad_norms):.1e}"
            })
            if i == len(train_loader) - 1 and old_criterion is not None and new_criterion is not None:
                loss_transition_alpha = min(1.0, loss_transition_alpha + 0.2)
                print(f"损失函数混合比例: {loss_transition_alpha:.1f}")
                
        train_loss /= len(train_subset)
        
        # 验证阶段
        model.eval()
        val_loss = 0.0
        val_dice = 0.0
        val_iou = 0.0
        val_preds = []      
        val_targets = []
        
        with torch.no_grad():
            # 可视化第一个样本
            if epoch % 2 == 0:  # 每隔2个epoch可视化一次
                image, mask = val_subset[0]
                image = image.unsqueeze(0).to(DEVICE)
                pred = model(image)
                
                pred_sigmoid = torch.sigmoid(pred)
                pred_bin = (pred_sigmoid > 0.5).float()
                
                print(f"预测值范围: {pred_sigmoid.min().item():.4f}-{pred_sigmoid.max().item():.4f}")
                print(f"真实前景像素: {mask.sum().item()}, 预测前景像素: {pred_bin.sum().item()}")
                
                # 可视化
                fig, axs = plt.subplots(1, 3, figsize=(12, 4))
                axs[0].imshow(image.squeeze().cpu(), cmap='gray'); axs[0].set_title("原图")
                axs[1].imshow(mask.squeeze(), cmap='gray'); axs[1].set_title("真实Mask")
                axs[2].imshow(pred_bin.squeeze().cpu(), cmap='gray'); axs[2].set_title("预测Mask")
                plt.show()
        
        with torch.no_grad():
            for images, masks in val_loader:
                images = images.to(DEVICE)
                masks = masks.to(DEVICE)
                
                outputs = model(images)
                outputs = torch.sigmoid(outputs)
                
                val_loss += criterion(outputs, masks).item() * images.size(0)
                
                # 记录用于后续自动阈值搜索
                val_preds.append(outputs.cpu())
                val_targets.append(masks.cpu())
                
                # 计算Dice和IoU
                preds = (outputs > 0.5).float()
                val_dice += dice_coeff(outputs, masks).cpu().item()
                val_iou += iou_score(outputs, masks).cpu().item()
        
        
        # 自动阈值搜索
        val_preds = torch.cat(val_preds, dim=0)
        val_targets = torch.cat(val_targets, dim=0)
        best_thresh, best_dice_val = find_best_threshold(val_preds, val_targets)
        
        preds_bin = (val_preds > best_thresh).float()
        val_dice = dice_coeff(preds_bin, val_targets).item()
        val_iou = iou_score(preds_bin, val_targets).item()
        
        val_loss /= len(val_subset)
        
        print(f"Epoch {epoch+1} 最佳阈值: {best_thresh:.2f} | 最佳Dice: {best_dice_val:.4f}")
        
        # 保存最佳模型并检查早停
        if best_dice_val > best_dice + 0.01:
            best_dice = best_dice_val
            torch.save({
                'epoch': epoch,
                'stage': stage,
                'model_state_dict': model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'loss': val_loss,
                'dice': best_dice_val,
                'threshold': best_thresh
            }, fold_checkpoint_path)
            patience_counter = 0  # 重置计数器
            print(f"第 {fold+1} 折模型已保存 - Epoch {epoch+1}, Dice: {best_dice_val:.4f}")
        else:
            patience_counter += 1
            if patience_counter >= PATIENCE:
                print(f"第 {fold+1} 折早停触发：{PATIENCE}个 epoch 没有提升")
                break
            print(f"第 {fold+1} 折当前Dice未提升，计数器：{patience_counter}/{PATIENCE}")
        
        # 打印训练信息
        print(f"第 {fold+1} 折 Epoch {epoch+1}/{config.total_epochs} (阶段 {stage}/3)")
        print(f"训练损失: {train_loss:.4f} | 验证损失: {val_loss:.4f}")
        print(f"Dice: {val_dice:.4f} | IoU: {val_iou:.4f} | 最佳Dice: {best_dice_val:.4f}")
        print(f"当前学习率 - Encoder: {optimizer.param_groups[0]['lr']:.2e} | Decoder: {optimizer.param_groups[1]['lr']:.2e}")
        print("=" * 60)
    
    # 训练结束后加载最佳模型进行评估
    print(f"\n===== 第 {fold+1} 折训练结束，加载最佳模型进行评估 =====")
    
    # 加载最佳模型
    checkpoint = torch.load(fold_checkpoint_path, weights_only=True)
    model.load_state_dict(checkpoint['model_state_dict'])
    
    # 获取保存的最佳阈值
    best_thresh = checkpoint.get('threshold', 0.5)
    print(f"加载最佳模型 (Epoch {checkpoint['epoch']+1}), Dice: {checkpoint['dice']:.4f}, 阈值: {best_thresh:.2f}")
    
    # 在测试集上评估
    test_dice, test_iou, visual_samples = evaluate_model(model, test_loader, best_thresh, DEVICE)
    print(f"测试集性能 - Dice: {test_dice:.4f}, IoU: {test_iou:.4f}")
    
    # 可视化样本
    for i, (image, mask, pred) in enumerate(visual_samples):
        fig, axs = plt.subplots(1, 3, figsize=(15, 5))
        
        # 原始图像
        axs[0].imshow(image, cmap='gray')
        axs[0].set_title(f"测试样本 {i+1} - 原图")
        axs[0].axis('off')
        
        # 真实掩码
        axs[1].imshow(mask, cmap='gray')
        axs[1].set_title(f"真实掩码 (前景像素: {mask.sum():.0f})")
        axs[1].axis('off')
        
        # 预测掩码
        axs[2].imshow(pred, cmap='gray')
        axs[2].set_title(f"预测掩码 (前景像素: {pred.sum():.0f})")
        axs[2].axis('off')
        
        plt.tight_layout()
        plt.savefig(f"fold{fold}_sample{i}.png")
        plt.show()
        
    # 保存测试结果
    results_path = f"fold{fold}_results.json"
    
    # 保存测试结果
    test_results = {
        'fold': int(fold),
        'best_epoch': int(checkpoint['epoch']),
        'best_val_dice': float(checkpoint['dice']),
        'best_threshold': float(best_thresh),
        'test_dice': float(test_dice),
        'test_iou': float(test_iou),
        'visual_samples': [str(sample[0].shape) for sample in visual_samples]  # 只保存形状信息
    }
    
    # 将结果添加到交叉验证结果列表
    cv_results.append(test_results)
    
    # 保存结果到文件
    results_path = f"fold{fold}_results.json"
    with open(results_path, 'w') as f:
        json.dump(test_results, f, indent=4)
    print(f"测试结果已保存到 {results_path}")

# 所有折完成后汇总结果
print("\n===== 交叉验证最终结果汇总 =====")
avg_val_dice = np.mean([r['best_val_dice'] for r in cv_results])
avg_test_dice = np.mean([r['test_dice'] for r in cv_results])
avg_test_iou = np.mean([r['test_iou'] for r in cv_results])

print(f"平均验证集Dice: {avg_val_dice:.4f}")
print(f"平均测试集Dice: {avg_test_dice:.4f}")
print(f"平均测试集IoU: {avg_test_iou:.4f}")

# 绘制Dice分布图
plt.figure(figsize=(10, 6))
folds = [f"Fold {i+1}" for i in range(len(cv_results))]
val_dices = [r['best_val_dice'] for r in cv_results]
test_dices = [r['test_dice'] for r in cv_results]

x = np.arange(len(folds))
width = 0.35

plt.bar(x - width/2, val_dices, width, label='验证集Dice')
plt.bar(x + width/2, test_dices, width, label='测试集Dice')

plt.xlabel('交叉验证折')
plt.ylabel('Dice系数')
plt.title('各折验证集与测试集Dice系数对比')
plt.xticks(x, folds)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.ylim(0, 1.0)

# 添加数值标签
for i, v in enumerate(val_dices):
    plt.text(i - width/2, v + 0.02, f"{v:.3f}", ha='center')
for i, v in enumerate(test_dices):
    plt.text(i + width/2, v + 0.02, f"{v:.3f}", ha='center')

plt.tight_layout()
plt.savefig("cv_dice_comparison.png")
plt.show()    
```
